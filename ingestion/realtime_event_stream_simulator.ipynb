```python
import pandas as pd
import time
from pathlib import Path
```

```python
PROJECT_ROOT = Path(r"D:\soc-dashboard-suite-main\soc-dashboard-suite-main")

EMAIL_EVENTS_PATH = PROJECT_ROOT / "data" / "normalized" / "email_events_normalized.csv"
STREAM_OUTPUT_PATH = PROJECT_ROOT / "data" / "stream" / "email_event_stream.csv"

email_df = pd.read_csv(EMAIL_EVENTS_PATH, parse_dates=["event_time"])

email_df.sort_values("event_time", inplace=True)
email_df.reset_index(drop=True, inplace=True)

email_df.head()
```

```python
# Create empty stream output file
stream_df = pd.DataFrame(columns=email_df.columns)
STREAM_OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
# Write header once before streaming starts
email_df.iloc[0:0].to_csv(STREAM_OUTPUT_PATH, index=False)

print("Initialized stream with header")

```

```python
for i in range(len(email_df)):
    event = email_df.iloc[[i]]

    event.to_csv(STREAM_OUTPUT_PATH, mode="a", header=False, index=False)

    print(f"Streamed event {i+1}/{len(email_df)}")

    time.sleep(0.1)  # simulate delay between events
```
