# Identity Resolution Layer

## Objective
Map email identities to internal user IDs for cross-signal correlation.


```python
import pandas as pd
from pathlib import Path
import random
```

```python
PROJECT_ROOT = Path(r"D:\soc-dashboard-suite-main\soc-dashboard-suite-main")

EMAIL_ALERTS_PATH = PROJECT_ROOT / "data" / "enriched" / "alerts_scored_prioritized.csv"
OUTPUT_PATH = PROJECT_ROOT / "data" / "enriched" / "identity_directory.csv"

email_df = pd.read_csv(EMAIL_ALERTS_PATH)
```

```python
email_users = email_df["recipient_email"].dropna().unique()
auth_users = [f"user{i}" for i in range(1, 51)]

directory = pd.DataFrame({
    "recipient_email": email_users,
    "user_id": [random.choice(auth_users) for _ in email_users]
})

directory.head()
```

```python
directory.to_csv(OUTPUT_PATH, index=False)
print("Identity directory saved to:", OUTPUT_PATH)
```
